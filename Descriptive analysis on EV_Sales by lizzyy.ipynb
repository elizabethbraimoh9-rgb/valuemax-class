```python
#checking missing values#
missing_values =df.isnull().sum()
print(missing_values)
```

```python
#visualise missing values#
plt.figure(figsize=(10,5))
sns.heatmap(df.isnull(),
            cbar=False,
cmap='viridis')
plt.title('Missing values Heatmap')
plt.show()
```

```python
df.copy()
```

```python
#replacin NaN value to 0
df['EV_Sales_Quantity'] =df['EV_Sales_Quantity'].fillna(0)
df['Vehicle_Category'] =df['Vehicle_Category'].fillna(0)
df['Vehicle_Type'] =df['Vehicle_Type'].fillna(0)
```

```python
#check duplicate#
print("Duplicate Rows:",
df.duplicated().sum())
```

###Data cleaning###

```python
#convert date column to datetime#
if 'Date' in df.columns:
    df['Date']
pd.to_datetime(df['Date'])
```

```python
#dropping none value#
df.dropna()
```

```python
#QUESTION.1#
#OVERALL ELECTRIC VEHICLES SALES TREND OVER TIME#
yearly_sales =df.groupby('Year')['EV_Sales_Quantity'].sum().reset_index()

plt.figure(figsize=(12,6))
sns.lineplot(data=yearly_sales,
             x='Year',
             y='EV_Sales_Quantity')
plt.title('overall EV Sales Trend Over Time')
plt.xlabel('Year')
plt.ylabel('EV Sales Quantity')
plt.grid(True)
plt.show()
             
```

```python
#QUESTION.2#
#STATES WITH HIGHEST EV ADOPTION#
state_sales = df.groupby('State')['EV_Sales_Quantity'].sum().sort_values(
ascending=False).reset_index()
plt.figure(figsize=(14,8))
sns.barplot(data=state_sales.head(10),
            x='EV_Sales_Quantity',
            y='State',
            palette='viridis')
plt.title('Top 10 States with Highest EV Adoption')
plt.xlabel('EV Sales Quantity')
plt.ylabel('State')
plt.show()
```

```python
#QUESTION 3#
#VEHICLES CATEGORIES DOMINATING EV SALES#
category_sales=df.groupby('Vehicle_Category')['EV_Sales_Quantity'].sum().reset_index()
plt.figure(figsize=(10,6))
plt.pie(category_sales['EV_Sales_Quantity'],
labels=category_sales['Vehicle_Category'],
        autopct='%1.2f%%')                     #shows percentage in 2 decimal place#
plt.title('EV Sales Distribution by Vehicles Category')
plt.show()
```

```python
#QUESTION 4#
#VEHICLE CLASS DISTRIBUTION#
class_sales =df.groupby('Vehicle_Class')['EV_Sales_Quantity'].sum().reset_index()
plt.figure(figsize=(10,4))
sns.barplot(data=class_sales.head(10),
            x='Vehicle_Class',
            y='EV_Sales_Quantity',
            palette='magma')
plt.title('EV Sales by Vehicle Class')
plt.xticks(rotation=40)
plt.show()
```

```python
#QUESTION 5#
#MONTHLY EV SALES TREND#
monthly_sales=df.groupby('Month_Name')['EV_Sales_Quantity'].sum().reset_index()
#display result
monthly_sales
```

```python
month_order = [
    'jan', 'feb', 'mar', 'apr', 'may', 'jun',
    'jul', 'aug', 'sep', 'oct', 'nov', 'dec'
]

# Convert Month_Name to ordered categorical
monthly_sales['Month_Name'] = pd.Categorical(
    monthly_sales['Month_Name'],
    categories=month_order,
    ordered=True
)

# Sort from Jan to Dec
monthly_sales = monthly_sales.sort_values('Month_Name')

print(monthly_sales)
```

```python
#visualise
plt.figure(figsize=(10,5))
plt.plot(monthly_sales['Month_Name'],
         monthly_sales['EV_Sales_Quantity'],
         marker='o')

plt.title('Monthly EV Sales Trend')
plt.xlabel('Month')
plt.ylabel('EV Sales Quantity')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
#QUESTION 6#
#TOP VEHICLES TYPE#
vehicle_type_sales =df.groupby('Vehicle_Type')['EV_Sales_Quantity'].sum().sort_values(
    ascending=False).reset_index()
plt.figure(figsize=(14,8))
sns.barplot(data=vehicle_type_sales.head(10),
            x='EV_Sales_Quantity',
            y='Vehicle_Type',
            palette='coolwarm')
plt.title('Top Vehicle by EV Sales')
plt.show()
```

```python
#QUESTION 7#
#SALES DISTRIBUTION#
plt.figure(figsize=(10,6))
sns.histplot(df['EV_Sales_Quantity'],
             bins=30,
             kde=False)
plt.title('Distribution of EV Sales Quantity')
plt.show()
```

```python
#QUESTION 8#
#STATE VS VEHICLES CATEGORY#
heatmap_data=df.pivot_table(values='EV_Sales_Quantity',index='State',
columns='Vehicle_Category',
aggfunc='sum')
plt.figure(figsize=(14,10))
sns.heatmap(heatmap_data,
            cmap='coolwarm')           #the heatmap highlights category dominance across states
plt.title('State vs Vehicle Category EV Sales')
plt.show()
```

```python
#QUESTION 9#
#YEARLY SALES BY VEHICLE CATEGORY#
Year_category=df.groupby(['Year','Vehicle_Category'])['EV_Sales_Quantity'].sum().reset_index()
plt.figure(figsize=(14,7))
sns.lineplot(data=Year_category,            #helps to knw growth pattern across categories#
             x='Year',
             y='EV_Sales_Quantity',
             hue='Vehicle_Category')
plt.title('Yearly EV Sales by Vehicle Category')
plt.show()
```

```python
#QUESTION 10#
#LOW OR ZERO ADOPTIVE ANALYSIS#
low_sales=df[df['EV_Sales_Quantity'] <=5]
print("Low Adoption Records:",
low_sales.shape[0])
df.dropna()
low_sales.head()
```

```python
#correlation analysis#
numeric_cols=df.select_dtypes(include=np.number)
correlation= numeric_cols.corr()
plt.figure(figsize=(8,6))
sns.heatmap(correlation,
            annot=True,
            cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
